Step 2: Import packages and setup the face detector and the face alignment model.

```
import dlib
import cv2
try:
  from google.colab.patches import cv2_imshow
except:
  print("Current runtime not on Google Colab.")
  print("Images will not be able to show within notebook.")
import importlib
fa = importlib.import_module('.image_processing', package='2020-COMP7404A-GroupB') # fa for Face Alignment
models = {}
models['detector'] = dlib.get_frontal_face_detector() 
models['predictor'] = dlib.shape_predictor('shape_predictor_68_face_landmarks.dat')
```

## For step 3, you have 2 choices: 
(a) Face alignment with image URL \\
(b) Face alignment of your own image uploaded 

Step 3a: With any image URL, you can detect and align the faces in the image.

```
# Example image from FFHQ Dataset
# 
# A Style-Based Generator Architecture for Generative Adversarial Networks
# Tero Karras (NVIDIA), Samuli Laine (NVIDIA), Timo Aila (NVIDIA)
# https://arxiv.org/abs/1812.04948
#
# https://github.com/NVlabs/ffhq-dataset

# Replace the URL with any image URL you like!
imageURL = 'https://github.com/NVlabs/ffhq-dataset/raw/master/ffhq-teaser.png'

fa.downloadImage(imageURL,output_file='test.jpg')
img = fa.faceAlignmentFromFile(f='test.jpg',models=models)

# cv2.imwrite('test_out.jpg',img) # write the output image if you want to
```
